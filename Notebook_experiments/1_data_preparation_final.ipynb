## 1) **Problem Statement**
- This study aims to develop a machine learning model that predicts the time it takes for graduates to secure employment after graduation. Using tracer study data, the research analyzes how academic, experiential, and behavioral factors influence employability outcomes.


## 2) **Data Collection**
- The dataset was obtained from the Dean of the Institute of Education in the form of a consolidated Excel file for analysis. Only non-identifiable academic data were included, with all personal identifiers removed to ensure privacy.

- All records were coded and securely stored with restricted access limited to the researcher. Confidentiality was strictly maintained, and no personally identifiable information was disclosed, in compliance with institutional ethical standards and the Data Privacy Act of 2012.

- The data originally consists of 87 columns and 158 rows.

#### **Renaming Variables**

```python
# ---------------------------------
# Rename Variables
# ---------------------------------

df = df.rename(columns={

    # Target
    'H5.  How long did it take you to find your current job after your graduation?':
        'time_to_employment',
    
    'H2.  How many jobs (including your current situation) have you had altogether since graduation?':
        'employment_status',
    
    # Academic
    'A0. What program did you complete?':
        'program',

    'A4.   On the average, how many hours per week did you spend on study activities OUTSIDE of classes during the course of your study?  (This includes doing class requirements, participating in co- and extra-curricular activities.)':
        'study_hours',

    # Internship
    'B1.  Did you do any internship (e.g., student teacher, etc.) during your course of studies (this does not refer to team projects, practical courses, etc.)?':
        'internship',

    'B2.  How many MANDATORY internships did you do in total  during your course of studies?':
        'mandatory_internship_num',

    # Working Student
    'C1.  Were you employed during your study? (Please include full-time as well as part-time work; do NOT include internships.)':
        'working_student',

    # Learning Environment
    'D1.  To what extent were the following aspects of teaching and learning emphasized in your studies? [Internship and practical learning]':
        'experiential_learning',

    'D1.  To what extent were the following aspects of teaching and learning emphasized in your studies? [Project and/or problem -oriented learning]':
        'project_based_learning',

    # Skills
    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Ability to adapt to changing conditions]':
        'adaptability',

    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Analytical thinking]':
        'analytical_thinking',

    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Ability to organize my work processes efficiently ]':
        'work_organization',

    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Ability to work productively with others ]':
        'teamwork',

    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Ability to perform well under pressure]':
        'stress_management',

    'E1.  To what extent did you acquire the following skills/competencies upon graduation? [Mastery of my field/subject-specific knowledge]':
        'applied_knowledge',

    'E3.   In retrospect, to what extent are you satisfied with your studies in general?':
        'study_satisfaction',

    # Job Search
    'F3.  When did you start searching for a job?':
        'search_timing',

    'F4.  How many employers did you approach for your first employment after completion of your study program?':
        'num_applications',

    'F6.  How many employers contacted you for exams or interview? ':
        'num_interviews',

    'F1.  In the FIRST SIX MONTHS after graduation, did you search for a job? ':
        'job_search',    

    # Demographics
    'M1.  What is your gender?':
        'sex',

    'M2.  In which year were you born?':
        'birth_year'

})
```

> **Note:** B1, F1, and H2 served as filter questions used to implement survey skip logic and determine eligibility for follow-up items. However, **B1 was also retained as a predictor variable** in the analysis because participation in internships may be associated with graduate employment outcomes. In contrast, **F1 and H2 were used solely to establish the applicability of subsequent questions and were not included as predictor nor target variables.**

#### **Handling Skip Logic and Missing Values**

```python
# ---------------------------------
# Handling H5 Missing Values in H5
# ---------------------------------

mask = (
    (df['employment_status'] == 'No job since graduation') &
    (df['time_to_employment'].isna())
)

df.loc[mask, 'time_to_employment'] = 'No job since graduation'

# ---------------------------------
# Handle Missing Values in B2
# ---------------------------------

mask = (
    (df['internship'] == 'No') &
    (df['mandatory_internship_num'].isna())
)

df.loc[mask, 'mandatory_internship_num'] = '5. Zero'

# ---------------------------------
# Handle Missing Values in F3, F4, and F6
# ---------------------------------

mask = (
    (df['job_search'] == 'No')
)

df.loc[
    mask & df['search_timing'].isna(),
    'search_timing'
] = 'Not Applicable'

df.loc[
    mask & df['num_applications'].isna(),
    'num_applications'
] = 'Not Applicable'

df.loc[
    mask & df['num_interviews'].isna(),
    'num_interviews'
] = 'Not Applicable'
```

### **Variable Encoding**

```python
# ---------------------------------
# Likert Scale Variables
# ---------------------------------

likert_map = {

    'To no extent': 1,
    'To a little extent': 2,
    'To some extent': 3,
    'To a large extent': 4,
    'To a very large extent': 5

}

likert_columns = [

    # Skills
    'adaptability',
    'analytical_thinking',
    'work_organization',
    'teamwork',
    'stress_management',
    'applied_knowledge',

    # Learning Environment
    'experiential_learning',
    'project_based_learning',
    'study_satisfaction'

]

for col in likert_columns:
    df[col] = df[col].map(likert_map)


# ---------------------------------
# Binary Variables
# ---------------------------------

binary_maps = {

    'internship': {
        'Yes': 1,
        'No': 0
    },

    'working_student': {
        'Yes': 1,
        'No': 0
    },

    'sex': {
        'Male': 0,
        'Female': 1
    }

}

for col, mapping in binary_maps.items():
    df[col] = df[col].map(mapping)


# ---------------------------------
# Ordinal Variables
# ---------------------------------

study_hours_map = {

    '10 hours or less':1,
    '11 to 19 hours':2,
    '20 to 29 hours':3,
    '30 to 39 hours':4,
    '40 hours or more':5

}

mandatory_internship_map = {

    '5. Zero':0,
    '1.  One':1,
    '2.  Two':2,
    '3.  Three':3,
    '4.  Four or more':4

}

search_timing_map = {

    'Prior to graduation':1,
    'Right after graduation':2,
    'One to less than three months after graduation':3,
    'Three to less than six months after graduation':4,
    'Six months or more after graduation':5,
    'Never applied for a job':6,
    'Not Applicable':7

}

applications_map = {

    'I have not approached an employer':0,
    'one employer':1,
    'two to four employers':2,
    'five to nine employers':3,
    '10 or more employers':4,
    'Not Applicable':5

}

interviews_map = {

    'No employer acknowledged my application':0,
    'From one employer':1,
    'From two to four employers':2,
    'From five to nine employers':3,
    'From 10 or more employers':4,
    'Not Applicable':5

}

df['study_hours'] = df['study_hours'].map(study_hours_map)

df['mandatory_internship_num'] = df['mandatory_internship_num'].map(
    mandatory_internship_map
)

df['search_timing'] = df['search_timing'].map(
    search_timing_map
)

df['num_applications'] = df['num_applications'].map(
    applications_map
)

df['num_interviews'] = df['num_interviews'].map(
    interviews_map
)


# ---------------------------------
# Nominal Variable (One-Hot Encoding)
# (Temporary Disabled)
# ---------------------------------

ENCODE_PROGRAM = False

if ENCODE_PROGRAM:
    df = pd.get_dummies(
    
        df,
    
        columns=['program'],
    
        prefix='program',
    
        drop_first=True,
    
        dtype=int
    
    )

#ENCODE_PROGRAM = True
#Note: When ready to apply
#Encode the variables immediately before model training.
```

```python
# ---------------------------------
# TESTING: Verify Encoded Variables
# ---------------------------------

print("="*60)
print("ENCODED DATASET OVERVIEW")
print("="*60)

print(f"\nRows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

# ---------------------------------
# Preview Dataset
# ---------------------------------

print("\nFirst 5 Rows:")
display(df.head())

# ---------------------------------
# Data Types
# ---------------------------------

print("\nData Types:")
display(df.dtypes.to_frame(name='Data Type'))

# ---------------------------------
# Missing Values
# ---------------------------------

print("\nMissing Values:")
display(df.isnull().sum().to_frame(name='Missing Count'))

# ---------------------------------
# Verify Encoded Variables
# ---------------------------------

encoded_columns = [

    'study_hours',
    'internship',
    'mandatory_internship_num',
    'working_student',
    'adaptability',
    'analytical_thinking',
    'work_organization',
    'teamwork',
    'stress_management',
    'applied_knowledge',
    'experiential_learning',
    'project_based_learning',
    'study_satisfaction',
    'search_timing',
    'num_applications',
    'num_interviews',
    'sex'

]

print("\nUnique Values After Encoding:")

for col in encoded_columns:

    print(f"\n{col}")
    print(sorted(df[col].dropna().unique()))

# ---------------------------------
# Summary Statistics
# ---------------------------------

print("\nSummary Statistics:")
display(df.describe(include='all'))
```

```python
# ---------------------------------
# TESTING: Verify Composite Indices
# ---------------------------------

print("="*60)
print("COMPOSITE INDICES")
print("="*60)

composite_cols = [
    'skills_index',
    'learning_environment_index'
]

# Check if indices exist
print("\nComposite Variables Created:")
print(df[composite_cols].head())

# Data types
print("\nData Types:")
print(df[composite_cols].dtypes)

# Missing values
print("\nMissing Values:")
print(df[composite_cols].isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
display(df[composite_cols].describe().round(2))

# ---------------------------------
# TESTING: Recompute Skills Index
# ---------------------------------

skills_check = df[[
    'adaptability',
    'analytical_thinking',
    'work_organization',
    'teamwork',
    'stress_management',
    'applied_knowledge'
]].mean(axis=1)

print("Skills Index Verification:")

if np.allclose(skills_check, df['skills_index'], equal_nan=True):
    print("✅ Skills Index computed correctly.")
else:
    print("❌ Skills Index contains inconsistencies.")

# -------------------------------------------------
# TESTING: Recompute Learning Environment Index
# -------------------------------------------------

learning_check = df[[
    'experiential_learning',
    'project_based_learning',
    'study_satisfaction'
]].mean(axis=1)

print("\nLearning Environment Index Verification:")

if np.allclose(learning_check,
               df['learning_environment_index'],
               equal_nan=True):
    print("✅ Learning Environment Index computed correctly.")
else:
    print("❌ Learning Environment Index contains inconsistencies.")

# -------------------------------------------------
# TESTING: Verify the range
# -------------------------------------------------

print("\nRange Check")

for col in composite_cols:

    print(f"\n{col}")
    print(f"Minimum: {df[col].min():.2f}")
    print(f"Maximum: {df[col].max():.2f}")

    if df[col].between(1, 5).all():
        print("✅ Values are within the expected range (1–5).")
    else:
        print("❌ Some values fall outside the expected range.")
```

### **Age**

```python
CURRENT_YEAR = 2025

df['age'] = CURRENT_YEAR - df['birth_year']
```

> **Note:** Age was computed by subtracting the respondent's birth year from 2025, which served as the reference year for the graduate tracer survey. A fixed reference year was used to ensure that age was measured consistently for all respondents during the study period and would not vary based on the year in which the dataset was analyzed.

```python
# ---------------------------------
# TESTING: Verify Age Variable
# ---------------------------------

print("="*60)
print("AGE VERIFICATION")
print("="*60)

# Preview
print("\nBirth Year and Age:")
display(df[['birth_year', 'age']].head())

# Data type
print("\nData Types:")
print(df[['birth_year', 'age']].dtypes)

# Missing values
print("\nMissing Values:")
print(df[['birth_year', 'age']].isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
display(df[['age']].describe().round(2))

# ---------------------------------
# TESTING: Verify the Computation
# ---------------------------------

# Recalculate age
age_check = 2025 - df['birth_year']

print("Age Calculation Verification:")

if age_check.equals(df['age']):
    print("✅ Age calculated correctly.")
else:
    print("❌ Age calculation contains inconsistencies.")

# ---------------------------------
# TESTING: Sanity Check
# ---------------------------------

invalid_age = df[
    (df['age'] < 15) |
    (df['age'] > 80)
]

print("\nInvalid ages:", len(invalid_age))
```
